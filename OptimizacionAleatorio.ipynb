```python
import random
import matplotlib.pyplot as plt
import numpy as np

def funcion(x, y):
    return y - x - 2*x**2 - 2*x*y - y**2

# Valores iniciales
xl = -2
xu = 2
yl = 1
yu = 3

num_iteraciones = 40


mejor_solucion = None
mejor_valor = float('inf')

valores_x = []
valores_y = []
valores_funcion = []

for _ in range(num_iteraciones):
    x = random.uniform(xl, xu)
    y = random.uniform(yl, yu)
    
    valor = funcion(x, y)
    
    valores_x.append(x)
    valores_y.append(y)
    valores_funcion.append(valor)
    
    if valor < mejor_valor:
        mejor_solucion = (x, y)
        mejor_valor = valor

print(f"{'x':^10}{'y':^10}{'funcion(x, y)':^20}")
for x, y, valor in zip(valores_x, valores_y, valores_funcion):
    print(f"{x:^10.4f}{y:^10.4f}{valor:^20.4f}")

# Gráfica de la función
x_vals = np.linspace(xl, xu, 100)
y_vals = np.linspace(yl, yu, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = funcion(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.scatter(mejor_solucion[0], mejor_solucion[1], mejor_valor, color='red', s=100, label='Mejor Solución')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('funcion(x, y)')
plt.legend()
plt.show()

print(f"\nLa mejor solución x = {mejor_solucion[0]}, y = {mejor_solucion[1]} con F(x,y)= {mejor_valor}.")
```
